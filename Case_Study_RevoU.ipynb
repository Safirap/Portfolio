#Import Library

# Read Data and Clean

```python
# read data from google sheet
sheet_url = 'https://docs.google.com/spreadsheets/d/10poofg-l8DMdtUgGy8mOpra2IHmA9EQC7drmF9AyYHA/edit#gid=1485085913'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df = pd.read_csv(sheet_url_trf)
df
```

```python
df.info()
```

```python
df.columns
```

```python
df_cleaned = df.copy()
df_cleaned['Release'] = pd.to_datetime(df_cleaned['Release'])

print(df_cleaned.dtypes)
df_cleaned.head()
```

```python
# delete NaN
df1 = df_cleaned.dropna()
df1
```

#Problem 1
Which game is the oldest and newest games in that dataset ?

```python
df1.sort_values(by=['Release','Name'],ascending=[True,False])
```

The oldest game is Hydlide. Release in 1984-12-01	
The Newest game is Crusader Kings. Release in 2020-09-01

#Problem 2
Which publisher published most of the games ?

```python
df1['Publisher'].value_counts().sort_values
```

The most publisher published is Electronic Arts 

#Problem 3
Which developer developed most of the games ?

```python
df1['Developer'].value_counts().sort_values
```

The most developer developed is Blizzard Entertainment

#Problem 4
Which series is the most sales ?

```python
df1[['Sales','Series']]
```

The most sales series is Minecraft with 33.0 Million

#Problem 5
Which series have the most games ?

```python
df1['Series'].value_counts()
```

The series that has the most games is StarCraft
